### Now, Applying name changes
* Change the first five columns names
* Change the contents of two columns

```python
# create a duplicate of the dataframe
df_backup = df.copy()
```

```python
# New Columns Names
col_names = {'USO2013': 'LandUseType', 'TU': 'UserType', 
             'DC': 'InletDiameter', 'M': 'GaugeVendor', 
            'UL': 'CurrentConsumption', }
df.rename(columns=col_names, inplace=True)
df.head(1)
```

```python
# Drop any row or column with fully missing values

print (df.shape)
# drop  any column with 100% nan (if any)
df.dropna(axis=1, how='all', inplace=True)
# drop  any row with 100% nan (if any)
df.dropna(axis=0, how='all', inplace=True)
# Check for shape changes
print (df.shape)
```

```python
# Drop any row if all the water consumption data are missing
data_columns = list(df)[4:]
df.dropna(axis=0, how='all', subset=data_columns, inplace=True)
df.shape
```

```python
print (f'Number of water consumption columns are:{len(data_columns)}')
```

```python
# Check the amount of rows corresponding to some threshold values

cols_thresh = [_ for _ in range(5, 86, 5)]
# equivalent remaining rows
equiv_rows = [df.dropna(axis=0, how='all', 
                        subset=data_columns, 
                        thresh=t).shape[0] for t in cols_thresh]
```

```python
num_rows = df.shape[0] # The number of rows before reduction
equiv_rows = np.array(equiv_rows) # Convert the results to a numpy array
# Get the percent of reduction
reduction = (num_rows - equiv_rows)/num_rows *100  
```

```python
plt.plot(cols_thresh, reduction,'ro');
plt.plot(cols_thresh, reduction,'b');
plt.title('Effect of drop NaN threshold on the percent of rows removed')
# plt.xlabel('If the count of columns with NaN > this amount, drop the row')
plt.xlabel('Keep the rows that contain at least [...] columns with valid numbers')
plt.ylabel('Percent of rows removed (%)')
plt.xticks(cols_thresh);
```

If we used the threshold = 80, we will lose only 15% of the rows, which I believe it is not bad.

```python
# Keeping the rows with at least 80 columns out of 85 columns with non-NaN values
# and dropping the rows with 5 or more columns with NaNs.

df.dropna(axis=0, how='all', 
                        subset=data_columns, 
                        thresh=80, inplace=True)
df.shape
```

```python
# How much NaNs in each column?

for col in list(df):
    print (f'Column: {col:<18} contains {df[col].isna().sum():5d} missing values')
```

It appears that most of the missing values are on the `CurrentConsumption` column followed by `GaugeVendor`, then the early 2009 months.

I will keep all the rows with NaN, and try to impute them later

***Imputing the values should be subject to time-series base, as filling missing February should be based on February of other years in addition to the trend if any.***

```python
# Replacing the Spanish symbols by English words for LandUseType and UserType columns

# LandUseType column
df['LandUseType'] = df['LandUseType'].map(uso_explain)

# UserType column
df['UserType'] = df['UserType'].map(tu_explain)

df.sample(10)
```

# Melting the data columns into three columns 
## The three columns should be Month, Year, Value

```python
# Taking a second backup of the dataframe
df_backup_2 = df.copy()
df.shape
```

```python
# Melting the dataframe to unpivot the data columns

new_df = pd.melt(df, id_vars=list(df)[:5], value_vars=list(df)[5:],
        var_name='MonthYear', value_name='Consumption')
print (new_df.shape)
new_df.sample(50)
```

```python
# The column MonthYear Needs to be split into Month, and Year.

df2=new_df.join(new_df['MonthYear'].
            str.split('_', 10, expand=True).
            rename(columns={0:'trash', 1:'Month', 2:'Year'}))
df2.sample(5)
```

```python
# Dropping the unnecessary columns

df2.drop(['MonthYear', 'trash'], axis=1, inplace=True)
print(df2.shape)
df2.head(2)
```

```python
#Convert the month to numeric values:

month_numbers = {'ENE': 1, 'FEB': 2, 'MAR': 3, 'ABR': 4, 
                 'MAY': 5, 'JUN': 6, 'JUL': 7, 'AGO': 8, 
                 'SEP': 9, 'OCT': 10, 'NOV': 11, 'DIC': 12}
df2['Month'] = df2['Month'].map(month_numbers)
df2.sample(5)
```

```python
# rename the column CurrentConsumption to RecentConsumption
df2.rename(columns={'CurrentConsumption': 'RecentConsumption'}, inplace=True)
```

```python
df2.sample(5)
```

```python
# Review the columns datatypes
```

**Problem in the `Year` column, it should be int**

```python
# Simple conversion

df2['Year'] = df2['Year'].astype(np.int64,  )
```

```python
# To impute this time series, we have to create a CurrentDate column, assuming the day is 15

df2['CurrentDate'] = pd.to_datetime(dict(year=df2.Year+2000, month=df2.Month, day=15))
df2.info(null_counts='all')
```
