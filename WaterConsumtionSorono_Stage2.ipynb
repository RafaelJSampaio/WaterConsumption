In this stage I have divided the data table to one main and three child tables that will be joined through primary-foriegn keys.
- The main table, I removed all text from it and placed foriegn keys instead
- The other three tables I have hot-encoded them, each with its suitable encoding

   - The `User` table was encoded based on the properties of its values.
   - The `Gauge` table was encoded based on teh one-hot-encoding method.

**Creating lighter version of the dataset**
*by creating a relational database for the textual columns
`Usage`, `User`, `Gauge`*

```python
# Creating a dataframe
Usage_df = pd.DataFrame(df.Usage.unique(), columns=['Usage_text']
            ).reset_index().rename(columns={'index':'Usage_id'})
# Creating a dictionary
Usage_dict = Usage_df.to_dict()['Usage_text']
# Reversing the dictionary
Usage_dict = {v:k for k, v in Usage_dict.items()}
Usage_dict
```

```python
# Creating a dataframe
User_df = pd.DataFrame(df.User.unique(), columns=['User_text']
            ).reset_index().rename(columns={'index':'User_id'})
# Creating a dictionary
User_dict = User_df.to_dict()['User_text']
# Reversing the dictionary
User_dict = {v:k for k, v in User_dict.items()}
User_dict
```

```python
# Creating a dataframe
Gauge_df = pd.DataFrame(df.Gauge.unique(), columns=['Gauge_text']
            ).reset_index().rename(columns={'index':'Gauge_id'})
# Creating a dictionary
Gauge_dict = Gauge_df.to_dict()['Gauge_text']
# Reversing the dictionary
Gauge_dict = {v:k for k, v in Gauge_dict.items()}
Gauge_dict
```

```python
# Substitute the IDs instead of the values in the main dataframe

# Creating IDs columns and mapping to the corresponding dictionaries
df = df.assign (Usage_ref=df['Usage'].map(Usage_dict), 
                User_ref=df['User'].map(User_dict),
                Gauge_ref=df['Gauge'].map(Gauge_dict))
```

# Encoding the text-columns to reflect ML properties

#### Creating some columns to reflect the properties of usage

```python
Usage_df = Usage_df.assign(industry=0)
# Usage_df['industry'] = np.where('industry' in Usage_df.Usage_text, 1, 0)
# Usage_df['industry'] = Usage_df['Usage_text'].apply(lambda x:1 if 'industry' in x else 0)
Usage_df['industry'] = pd.np.where(Usage_df.Usage_text.str.contains("Industry"), 1,0)
Usage_df
```

```python
# True if contains Industry or Infrastructure
Usage_df = Usage_df.assign(industry=0)

# True if contains Housing or Downtown
Usage_df = Usage_df.assign(housing=0)

# True if contains Parks or Vegetation
Usage_df = Usage_df.assign(agriculture=0)

# True if contains Downtown or Commerce
Usage_df = Usage_df.assign(commerce=0)

# True if contains Reserve 
Usage_df = Usage_df.assign(reserve=0)

# True if contains Infrastructure or Government or Parks
Usage_df = Usage_df.assign(government=0)

# True if contains Low 
Usage_df = Usage_df.assign(low_density=0)

# True if contains Medium 
Usage_df = Usage_df.assign(mid_density=0)

# True if contains High 
Usage_df = Usage_df.assign(high_density=0)

# Applying the conditions
use_str = Usage_df.Usage_text.str
Usage_df['industry'] = pd.np.where(use_str.contains("Industry"), 1,
                       pd.np.where(use_str.contains("Infrastructure"), 1,0))
Usage_df['housing'] = pd.np.where(use_str.contains("Housing"), 1,
                       pd.np.where(use_str.contains("Downtown"), 1,0))
Usage_df['agriculture'] = pd.np.where(use_str.contains("Parks"), 1,
                       pd.np.where(use_str.contains("Vegetation"), 1,0))
Usage_df['commerce'] = pd.np.where(use_str.contains("Downtown"), 1,
                       pd.np.where(use_str.contains("Commerce"), 1,0))
Usage_df['reserve'] = pd.np.where(use_str.contains("Reserve"), 1,0)
Usage_df['government'] = pd.np.where(use_str.contains("Government"), 1,
                       pd.np.where(use_str.contains("Infrastructure"), 1,
                       pd.np.where(use_str.contains("Parks"), 1,0)))
Usage_df['low_density'] = pd.np.where(use_str.contains("Low"), 1,0)
Usage_df['mid_density'] = pd.np.where(use_str.contains("Medium"), 1,0)
Usage_df['high_density'] = pd.np.where(use_str.contains("High"), 1,0)
Usage_df
```

```python
User_df
```

```python
# True if contains Industrial
User_df = User_df.assign(u_industry=0)

# True if contains Domestic
User_df = User_df.assign(u_domestic=0)

# True if contains Social 
User_df = User_df.assign(u_social =0)

# True if contains Commercial 
User_df = User_df.assign(u_commerce=0)

# True if contains Low 
User_df = User_df.assign(u_low_density=0)

# True if contains Medium or Normal
User_df = User_df.assign(u_mid_density=0)

# True if contains High 
User_df = User_df.assign(u_high_density=0)

# True if contains Residential  
User_df = User_df.assign(u_residential=0)


# Applying the conditions
use_str = User_df.User_text.str
User_df['u_industry'] = pd.np.where(use_str.contains("Industrial"), 1,0)
User_df['u_social'] = pd.np.where(use_str.contains("Social"), 1,0)
User_df['u_domestic'] = pd.np.where(use_str.contains("Domestic"), 1,0)
User_df['u_commerce'] = pd.np.where(use_str.contains("Commercial"), 1,0)
User_df['u_residential'] = pd.np.where(use_str.contains("Residential"), 1,0)
User_df['u_low_density'] = pd.np.where(use_str.contains("Low"), 1,0)
User_df['u_mid_density'] = pd.np.where(use_str.contains("Medium"), 1,
                       pd.np.where(use_str.contains("Normal"), 1,0))
User_df['u_high_density'] = pd.np.where(use_str.contains("High"), 1,0)
User_df
```

I found very few common properties that can be extracted.
I think it is better to number the gauges as G_GaugeName
This should bedone by 
1. changing all cases to SentenceCase
2. removing special characters like /, \, ", -, ., and so on...
3. replacing spaces by underscores


```python
Gauge_df = Gauge_df.assign(Gauge_names=Gauge_df.Gauge_text)
Gauge_df.Gauge_names = Gauge_df.Gauge_names.str.replace('1/2','half_in')
Gauge_df.Gauge_names = Gauge_df.Gauge_names.str.replace('3/4','triquarter_in')
Gauge_df.Gauge_names = Gauge_df.Gauge_names.str.replace('"',' in')
Gauge_df.Gauge_names = Gauge_df.Gauge_names.str.replace('-','')
Gauge_df.Gauge_names = Gauge_df.Gauge_names.str.replace('/','_')
Gauge_df.Gauge_names = Gauge_df.Gauge_names.str.replace(' ','_')
Gauge_df.Gauge_names = Gauge_df.Gauge_names.str.replace('__','_')
Gauge_df.Gauge_names = Gauge_df.Gauge_names.str.replace('__','_')
# Gauge_df.Gauge_names = Gauge_df.Gauge_names.str.lower()
# Gauge_df.Gauge_names = Gauge_df.Gauge_names.str.capitalize()
Gauge_df.Gauge_names = Gauge_df.Gauge_names.str.title()
# Gauge_df.Gauge_names = 'Gauge_' + Gauge_df.Gauge_names.astype(str)
Gauge_df
```

```python
# One_hot_encoding using pandas

Gauge_df2 = pd.concat([Gauge_df,pd.get_dummies(Gauge_df.Gauge_names, prefix='Gauge')], axis=1)
Gauge_df2
```
